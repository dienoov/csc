```python
import matplotlib.pyplot as plt
import numpy as np


def interpolate(x, y, xi):
    n = len(x)

    f = y.copy()
    for i in range(1, n):
        for j in range(n - 1, i - 1, -1):
            f[j] = (f[j] - f[j - 1]) / (x[j] - x[j - i])

    result = f[-1]
    for i in range(n - 2, -1, -1):
        result = result * (xi - x[i]) + f[i]

    return result

x = np.array([0.0, 1.0, 2.0])
y = np.array([2.0, 3.0, 7.0])

xi = 1.5
yi = interpolate(x, y, xi)

x_plot = np.arange(np.min(x), np.max(x) + 0.1, 0.1)
y_plot = [interpolate(x, y, xi) for xi in x_plot]

plt.plot(x_plot, y_plot)
plt.scatter(xi, yi)
plt.title("(x, y) = ({}, {})".format(xi, yi))
plt.grid()
plt.show()
```
