```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0.0, 1.0, 2.0])
y = np.array([1.0, 3.0, 2.0])

xi = 1.5
yi = np.interp(xi, x, y)

plt.plot(x, y)
plt.scatter(xi, yi)
plt.annotate("({}, {})".format(xi, yi), xy=(xi, yi), xytext=(xi + 0.05, yi + 0.05))
plt.grid()
plt.show()
```
